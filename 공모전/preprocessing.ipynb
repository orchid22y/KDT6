```python
# 모둘 로드
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
```

```python
# 데이터 불러오기
df=pd.read_excel('../data/시군구별 요양기관 현황(가공데이터).xlsx')
df
```

```python
# 결측치 많은 2008년도 데이터 삭제
df2 = df[df['C0'] != 2008]
df2
```

```python
# 결측치 많은 C18, C68 열 삭제

df3=df2.drop(columns=['C16','C18','C68'])
df3
```

```python
# C1 기준으로 지정된 순서 설정
custom_order = [
    '전국', '서울특별시', '부산광역시', '대구광역시', '인천광역시', '광주광역시',
    '대전광역시', '울산광역시', '세종특별자치시', '경기도', '강원특별자치도',
    '충청북도', '충청남도', '전북특별자치도', '전라남도', '경상북도', '경상남도',
    '제주특별자치도'
]

df3['C1'] = pd.Categorical(df3['C1'], categories=custom_order, ordered=True)

# C1을 지정된 순서로, C0을 오름차순으로 정렬
df3 = df3.sort_values(by=['C1', 'C0'])
df3
```

```python
df3.loc[df3['C1'] == '세종특별자치시']
```

```python
# C1이 '세종특별자치시'인 행의 C50~C61 열을 선형 보간
cols_to_interpolate = ['C50', 'C51', 'C52', 'C53', 'C54', 'C55', 'C56', 'C57', 'C58', 'C59', 'C60', 'C61']
df3.loc[df3['C1'] == '세종특별자치시', cols_to_interpolate] = df3.loc[df3['C1'] == '세종특별자치시', cols_to_interpolate].interpolate(method='linear', axis=0)

```

```python
df3.loc[df3['C1'] == '세종특별자치시']
```

```python
# C1이 '세종특별자치시'인 행의 C50~C61 KNN 알고리즘으로 결측치 채우기


cols_to_impute = ['C50', 'C51', 'C52', 'C53', 'C54', 'C55', 'C56', 'C57', 'C58', 'C59', 'C60', 'C61']
# '-' 값을 NaN으로 변환
df3[cols_to_impute] = df3[cols_to_impute].replace('-', np.nan)

# KNNImputer 객체 생성
imputer = KNNImputer(n_neighbors=5)

# '세종특별자치시'인 행에서 특정 열의 결측치를 KNN으로 채우기
df3.loc[df3['C1'] == '세종특별자치시', cols_to_impute] = imputer.fit_transform(df3.loc[df3['C1'] == '세종특별자치시', cols_to_impute])
```

```python
# 나머지 결측치 0으로 채우기

# '-' 값을 NaN으로 변환
df3 = df3.replace('-', np.nan)
df3.loc[:, df3.columns.difference(['C0', 'C1'])] = df3.loc[:, df3.columns.difference(['C0', 'C1'])].fillna(0)
```

```python
# DataFrame -> excel
df3.to_excel("preprocessing.xlsx",index=False)
```

```python
#C1 컬럼 원핫 인코딩
df3 = pd.get_dummies(df3, columns=['C1'], drop_first=True)
```

```python
# DataFrame -> excel
df3.to_excel("preprocessing_onehot.xlsx",index=False)
```
