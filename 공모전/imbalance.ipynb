```python
# 모듈로드
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
# 데이터 로드

df=pd.read_excel('../data/인구대비 의료 현황(가공데이터).xlsx')
df
```

```python
df2=df.drop(columns=df.columns[1:6])
df2
```

```python
# 선그래프 함수

def drawLineGraph(title,DF):
    sub=DF.columns.unique()[1:]
    area=['서울특별시', '부산광역시', '대구광역시', '인천광역시', '광주광역시', '대전광역시', '울산광역시',
       '세종특별자치시', '경기도', '강원도', '충청북도', '충청남도', '전라북도', '전라남도', '경상북도',
       '경상남도', '제주특별자치도']
    plt.figure(figsize=(16, 7))

    for s in sub:
        plt.plot(area,DF[s].values, label=s)
    
    plt.title(f'[{title}]')
    plt.xlabel('지역') 
    plt.ylabel('비율')
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.tight_layout()
    plt.savefig('img/' + title + '.png',dpi=100)
    plt.show()
    plt.close()
```

```python
drawLineGraph('인구대비 의료 자원 현황',df2)
```
