```python
# 모듈 로드
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
import folium
from folium import Choropleth
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from PIL import Image
import time
```

```python
# 데이터 불러오기
df=pd.read_excel('../data/의료수가와 변수(가공데이터).xlsx')
df
```

```python
df2 = df.set_index('구분')
df2
```

```python
df3=df2.T
df3
```

```python
# 상관관계 확인 하기
df3.corr()[7:]
```

```python
sub_list=df3.columns
sub_list
```

```python
sub_list=['의료인력(간호사 및 기타 인원)', '의료인력(의사,약사 등)','소비자물가지수(보건)', 'GDP', '인플레이션 전망', '고령화지수','요양기관현황']
```

```python
cost=['병원, 요양병원, 정신병원 및 종합병원', '의원', '치과의원 및 치과병원', '한의원 및 한방병원', '조산원','한국희귀·필수의약품센터', '보건진료소']
```

```python
# 산점도 그래프

fig, axs = plt.subplots(7,7,figsize=(25,35))

for idx,c in enumerate(cost):
    x=df3[c].values
    
    for i,sub in enumerate(sub_list):
        ax=axs[idx,i]
        y=[]

        y = df3[sub].values
        
        ax.scatter(x, y, marker='o')
        ax.set_xlabel(c)
        ax.set_ylabel(sub)
    #ax.set_title(f'{c} & {sub}')
plt.show()
```

```python
# 요양기관현황 기준 산점도 그래프

fig, axs = plt.subplots(7,7,figsize=(25,35))

for idx,c in enumerate(cost):
    x=df3[c].values
    
    for i,sub in enumerate(sub_list):
        ax=axs[idx,i]
        y=[]

        y = df3[sub].values
        
        ax.scatter(x, y, marker='o')
        ax.set_xlabel(c)
        ax.set_ylabel(sub)
    #ax.set_title(f'{c} & {sub}')
plt.show()
```

>>>> 병원 서비스권역 내 인구비율 시각화 하기

```python
# 데이터 불러오기
df=pd.read_excel('../data/7_4_병원 서비스권역 내 인구비율.xlsx')
df
```

```python
#필요 데이터 추출
df2020=df[(df['제공년도'] == '2020년') & (df['시군구'] == '전체')]
df2020=df2020[['제공년도', '시도', '값']]
df2020
```

```python
#필요 데이터 추출_2021년
df2021=df[(df['제공년도'] == '2021년') & (df['시군구'] == '전체')]
df2021=df2021[['제공년도', '시도', '값']]
df2021
```

```python
geo_path = '../data/SIDO_MAP_2022.json'

# 대한민국 중심 좌표 (예: 서울 좌표)
m2020 = folium.Map(location=[36.5, 127.5], zoom_start=7)

# Choropleth Layer 추가
Choropleth(
    geo_data=geo_path,
    data=df2020,
    columns=['시도', '값'],
    key_on='feature.properties.CTP_KOR_NM',
    fill_color='YlGnBu',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='병원 서비스권역 내 인구비율 (%)'
).add_to(m2020)

# 지도 표시
m2020.save("korea_population_coverage_map.html")
m2020
```

```python
# 대한민국 중심 좌표 (예: 서울 좌표)
m2021 = folium.Map(location=[36.5, 127.5], zoom_start=7)

# Choropleth Layer 추가
Choropleth(
    geo_data=geo_path,
    data=df2021,
    columns=['시도', '값'],
    key_on='feature.properties.CTP_KOR_NM',  # GeoJSON 파일의 지역명 필드명에 맞게 수정 필요
    fill_color='YlGnBu',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='병원 서비스권역 내 인구비율 (%)'
).add_to(m2021)

# 지도 표시
m2021.save("korea_population_coverage_map.html")
m2021
```
