```python
import pymysql
import pandas as pd

# 데이터베이스 연결
conn = pymysql.connect(
    host='172.20.49.39',
    user='ranyoung',
    password='1234',
    db='sql_project',
    charset='utf8')
# SQL 쿼리
query = '''
select p.country, p.property , g.gdp, d.density
from property38 as p
inner join population_density38 as d on p.country_id=d.Country_Code
inner join gdp38 as g on p.country_id=g.Country_Code;
'''
# 쿼리 결과를 DataFrame으로 읽어오기
maindf = pd.read_sql(query, conn)

# 연결 종료
conn.close()

# DataFrame 출력
print(maindf)
```

```python
import matplotlib.pyplot as plt
import koreanize_matplotlib
fig, ax = plt.subplots(figsize=(16, 8))

country =maindf['country']
value = maindf['property']
bar_labels = country

barlist=ax.bar(country, value, label=bar_labels, color='#b4ccd2')
barlist[0].set_color('#d55769')

ax.set_ylabel('가격 대비 소득 비율')
ax.set_title('OECD 부동산 가격 대비 소득 비율')
plt.xticks(range(38),country,rotation=80)
plt.savefig('img/' +'OECD 부동산 가격 대비 소득 비율(2022)' + '.png',dpi=100)
plt.show()
```

```python
import numpy as np
income = maindf['property'].to_list()
gdp = maindf['gdp'].to_list()
density=maindf['density'].to_list()
fig, axs = plt.subplots(1, 2, figsize=(16, 5))
sizes = np.array(maindf['property']) *70

c_list=['#954b63', '#7083a6']
for i, v in enumerate(['gdp', 'density']):
    ax = axs[i]
    if v == 'gdp':
        y = gdp
    else:
        y = density
    
    ax.scatter(income, y, s=sizes, alpha=0.5, c=c_list[i], edgecolors='w', linewidth=0.5)
    ax.set_xlabel('부동산 가격 대비 소득 비율')
    ax.set_ylabel(v)
    ax.set_title(f'{v} vs 부동산 가격 대비 소득 비율')	

plt.tight_layout()
plt.savefig('img/' +'소득비율 산점도' + '.png',dpi=100)
plt.show()
```
