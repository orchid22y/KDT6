#### [Matplotlib 활용한 시각화]
- matplotlib.pyplot 모듈을 사용
- 다양한 그래프 관련 함수들 & 클래스들 존재

```python
#[1] 모듈 로딩
    # 패키지명.모듈명
import matplotlib.pyplot as plt
```

```python
# [2] 그래프 그리기 - 선(line) 그래프
# plot(y축데이터)
plt.plot([1,2,3,4]) 
plt.xlabel("DATA")
plt.ylabel("RESULT")
plt.show()
```

```python
# [2] 그래프 그리기 - Line(선) 그래프
# plot(x축 데이터, y축 데이터)
plt.plot([10,20,30],[5,0,15])
```

```python
plt.show()
```

```python
# [2] 그래프 그리기 - Line(선) 그래프
# plot(x축 데이터, y축 데이터, fmf포멧스트링)
# fmf 포맷스트링 : 그래프 색상, 두께, 스타일 등등 지정
plt.plot([10,20,30],[5,0,15],'r--')
plt.show()
```

```python
# [2] 그래프 그리기 - Line(선) 그래프
# plot(x축 데이터, y축 데이터, fmf포멧스트링)
# fmf 포맷스트링 : 그래프 색상, 두께, 스타일 등등 지정
plt.plot([10,20,30],[5,0,15],'ro--')
plt.show()
```

```python
plt.plot([10,20,30],[5,0,15],'r^--')
plt.show()
```

```python
plt.plot([10,20,30],[5,0,15],'b<--')
plt.plot([10,17,21],[5,8,15],'k>--')
plt.show()
```

```python
plt.plot([10,20,30],[5,0,15],'b<--')
plt.plot([10,17,21],[5,8,15],'k>')
plt.show()
```

```python
# [2] 그래프 그리기 - Line(선) 그래프
# plot(x축 데이터, y축 데이터, line2D속성으로 설정)
# fmf 포맷스트링 : 그래프 색상 두께, 스타일 등등 지정
plt.plot([10,20,30],[5,0,15],color='red',marker='o',linestyle='dashed',markersize=10,linewidth=4)
plt.show()
```

```python
# [2] 그래프 그리기 - line(선) 그래프
#plt.plot([10,20,30],[5,25,35])
#plt.plot([10,20,30],[18,29,25])
#plt.plot([10,20,30],[25,25,25])
plt.plot([10,20,30],[15,25,15],'bo-',[10,20,30],[18,29,25],'ro--',[10,20,30],[25,25,25],'g<-')
```

```python
# [3] 그래프에 부가적인 기능 추가
# - 그래프 제목
# - X축, Y축 라벨
# - 여러개 그래프 경우 범례 => 그래프의 lavel속성값을 범례에서 사용함
plt.plot([10,20,30],[15,25,20],'bo--',label="V")
plt.plot([10,20,30],[5,30,28],'ro--',label="2")
plt.title('[MPLIB LINE2D]')
plt.xlabel("Data")
plt.ylabel('Result')
plt.legend()
plt.show()
```

```python
# [3] 그래프에 부가적인 기능 추가
# - 그래프 제목
# - X축, Y축 라벨
# - 여러개 그래프 경우 범례 => 그래프의 lavel속성값을 범례에서 사용함
#                               loc속성으로 위치 지정 가능 [기:best]
plt.plot([10,20,30],[15,25,20],'bo--',label="V")
plt.plot([10,20,30],[5,30,28],'ro--',label="2")
plt.title('[MPLIB LINE2D]')
plt.xlabel("Data")
plt.ylabel('Result')    
plt.legend(loc="center right")
plt.show()
```
