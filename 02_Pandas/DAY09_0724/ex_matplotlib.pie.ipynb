#### [ 산점도 그래프 ]
- 두개의 컬럼 (변수)의 관계를 시각화해서 표현
- 연속형(수치형) 데이터에 사용함

```python
# [1] 모듈 로딩
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
from matplotlib import font_manager as fm
from matplotlib import rc

# 적용한 폰트 파일
FONT_FILE=r'C:\Windows\Fonts\malgun.ttf'

# 폰트 패밀리 이름 가져오기
font_name=fm.FontProperties(fname=FONT_FILE).get_name()

# 새로운 폰트 패밀리 이름 지정
rc('font', family=font_name)
```

```python
# [2] 데이터 준비
F_DATA=r'C:\Users\KDP-27\Desktop\KDT6\EX_PANDAS06\DAY04_0716\auto_mpg.csv'
```

```python
# [3] CSV => DataFrame으로 저장
dataDF= pd.read_csv(F_DATA)
```

```python
# - 컬럼별 타입 확인
dataDF.dtypes
```

```python
# [4] 특정 컬럼의 데이터비율
#   -자동차 생산국가에 대한 비율
dataDF['origin'].astype('category')
```

```python
#  자동차 생산국가에 대한 비율
round(dataDF['origin'].value_counts()/dataDF['origin'].count(),2)
```

```python
#  자동차 생산국가에 대한 비율
originSR=dataDF['origin'].value_counts()
x=originSR.index
h=originSR.values
```

```python
#  자동차 생산국가에 대한 비율 시각화 -- bar그래프
originBar=plt.bar(x,h,color=['red','yellow','green'],edgecolor=['blue'])
# 막대에 라벨을 설정
plt.bar_label(originBar,h,label_type='center')

# x축 눈금 설정
plt.xticks([1,2,3],['US','EU','JP'],rotation=-45,fontsize=15)
plt.show()
```

```python
#  자동차 생산국가에 대한 비율 시각화 -- pie그래프
plt.pie(h,labels=['US','EU','JP'],autopct='%.1f%%')
plt.show()
```

```python
plt.pie(h,labels=['US','EU','JP'],autopct='%.1f%%',counterclock=False)
plt.show()
```

```python
plt.pie(h,labels=['US','EU','JP'],autopct='%.1f%%',
        startangle=90,explode=[0.1,0.1,0.1])
plt.show()
```

```python
plt.pie(h,labels=['US','EU','JP'],autopct='%.1f%%',
        startangle=90,wedgeprops={'linewidth':10,'width':0.25,'edgecolor':'y'})
plt.show()
```
