### [그룹화 처리]
- 주제 : 타이타닉에서 선실의 등급에 따른 생존 비율을 분석

```python
import pandas as pd
```

```python
DF=pd.read_csv('titanic.csv')
```

```python
DF.info()
```

```python
DF.duplicated().sum()
```

```python
DF.drop_duplicates(keep='first')
```

```python
DF2=DF[['survived','pclass']]
```

```python

Obj=DF2.groupby('pclass')
```

```python
resultDF=Obj.agg(['min','max','mean','median','sum'])
resultDF
```
