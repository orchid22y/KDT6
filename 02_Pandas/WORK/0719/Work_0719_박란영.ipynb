#### filter 메서드를 이용한 행과 열 선택
- filter 메서드 
* items : loc과 비슷한 방법으로 행 혹은 열을 선택   
* like : 행 또는 열 이름의 일부만 가지고 필터링  
* regex : 정규표현식을 이용하여 행 또는 열을 필터링
* axis : 행과 열을 선택 ( axis=0 은 행, axis=1은 열)

```python
# set_index 메서드를 이용하여 datetime열을 인덱스로 설정
df=df.set_index('datetime')
df
```

```python
# 시간이 00:00:00인 데이터만 필터링
df.filter(like='00:00:00',axis=0)
```

```python
# humidty와 Windspeed 열만 필터링
df.filter(items=['humidity','windspeed'],axis=1)
```

```python
# 열 이름 중에 'in'과 's'사이에 임의의 한 문자를 지니는 열
df.filter(regex='in.s') # 여기서 in과 s 사이의 "."은 숫자와 알파벳 등을 포함안 임의의 문자 1개
```

#### 데이터의 분위수를 구하는 quantile 메서드

```python
# Total_Review열의 데이터를 최소값을 0, 최대값을 1로 하였을 때 quantile 메서드와 for 반복문을 통해서 0.2단위로 각 분위수 구하고 그 값 출력
quantile=[0,0.2,0.4,0.6,0.8,1]

for idx in quantile:
    q=df['Total_Review'].quantile(idx,interpolation='lower')
    print(f'quantile({idx}) is {q}')
```

```python
# quantile 메서드에 전달되는 인자의 값이 커질수록 높은 분위수에 해당하며, 낮은 값일수록 낮은 분위수에 해당하는 값을 반환
```

```python
# interpolation : 분위수를 구할 때 특정 값 i와 j 사이에 구하고자 하는 분위수가 존재할 경우, 그 값을 어떻게 반환할 지를 결정하는 인자
# 디폴트값 : linear-선형 내삽을 통하여 분위수를 반환
# lower , higher : i와 j중에서 각각 작은, 큰 값을 반환
# nearest : i와 j중에서 보다 가까운 값을 반환
# midpoint : i와 j의 중간 값 반환
```

```python
for idx in quantile:
    q=df['Total_Review'].quantile(idx, interpolation='nearest')
    print(f'quantile({idx} is {q})')
```

#### 원하는 데이터만 필터링 하는 query메서드
- query : 필터링할 조건을 문자열 형태의 인자로 전달  
  - Boolean 인덱스 전달방법, loc나 iloc 메서드를 전달하는 방법에 비해 가독성이 좋고, 여러가지 조건을 and나 or로 동시에 적용할 때에 더욱 깔끔하게 코드를 작성할 수 있다.

```python
import seaborn as sns
```

```python
df=sns.load_dataset('penguins')
df.head()
```

  - query 메서드는 특정 수치형 변수 (열)의 대소비교나 부등식에 관한 >,>=,<,<=,==,!=를 사용할 수 있다
  - query 메서드 인자에는 문자열 형태로 '(변수명)(등호나부등호)(값)'형태로 전달하면 된다

```python
# 펭귄의 부리 길이 (bil_length)가 55mm 이상인 행만 필터링
df.query('bill_length_mm >= 55')
```

```python
# 위와 같은 결과를 boolean 인덱스를 전달하는 방법으로 필터링
df[df['bill_length_mm']>=55]
```

```python
# loc 사용
df.loc[df['bill_length_mm']>=55]
```

```python
# 부리의 길이가 55mm 이상이고 종이 'Gentoo'인 데이터 필터링
df.query('bill_length_mm >=55 and species=="Gentoo"' )
```

```python
# 외부 변수 참조 가능
lenght = 55
species='Gentoo'

df.query('bill_length_mm >= @lenght and species==@species')
```

- contains : 문자열 내 특정 문자열의 일부(알파벳의 조합 등)를 포함하는지 확인
- startswith, endswith : 특정 문자열의 일부로 시작하는지, 끝나는지 확인

```python
#contains 메서드를 이용하여 island 변수의 각 데이터 값이 'oe'를 포함하는 해만 필터링
df.query('island.str.contains("oe")',engine='python')
```

```python
# endswith 메서드를 사용하여 species변수에서 e로 끝나는 데이터들만 필터링
df.query('species.str.endswith("e")')
```

- isin : 특정 변수에서 리스트에 포함되어 있는 값들만 필터링

```python
# filtering이라는 변수에 "Adelie"와 "Chinstrap"을 리스트형으로 지정하고, penguins 데이터셋의 "species" 열에서 filtering변수에 저장한 값들만 필터링
filtering=["Adelie","Chinstrap"]
df.query('species.isin(@filtering)')
```

#### Pandas의 꽃, 그룹별 연산을 위한 groupby 메서드

```python
df=sns.load_dataset('titanic')
df.head()
```

```python
# 성별에 대한 생존율
df.groupby('sex')['survived'].mean()
```

```python
# 묶을 변수가 두개 이상이라면 groupby 메서드에 리스트형태로 그룹 할 인자 전달
# 좌석 등급 별 생존율

df.groupby(['sex','class'])['survived'].mean()
```

```python
#groupby를 통한 그룹 연산에서 mean메서드와 각 데이터의 길이를 구할 수 있는 count 메서드를 동시에 적용 가능
df.groupby(['sex','class'])['survived'].agg(['mean','count'])
```

```python
df.groupby(['sex','class'])[['survived','age']].agg({'survived':'mean'})
```

```python
def get_IQR(data):
    _3rd=data.quantile(.75)
    _1st=data.quantile(.25)
    return(np.abs(_3rd - _1st)*1.5)

df.groupby(['sex','class'])['age'].apply(get_IQR)
```
