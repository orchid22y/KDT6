```python
#연도별, 지역별 카메라대수
local_list=['강원','경기','경남','경북','광주','대구','대전','부산','서울','세종','울산','인천','전남','전북','제주','충남','충북']
count1=[]
for y in ['2020','2021','2022']:
    for l in local_list:
        year=cctvDF[cctvDF['설치연월'].str.contains(y, na=False)]
        idx=year[year['지역']==l].index
        total=year.loc[idx]['카메라대수'].sum()
        
        row = {'설치연도': y, '지역': l, '카메라대수': total}
        count1.append(row)

countDF2=pd.DataFrame(count1)
countDF2
```

```python
pivot_cctv_df = countDF2.pivot(index='지역', columns='설치연도', values='카메라대수')

pivot_cctv_df
```

```python
pivot_cctv_df.info()
```

```python
pivot_cctv_df2=pivot_cctv_df.copy()
for l in local_list:
        t=pivot_cctv_df2.loc[l].sum()
        pivot_cctv_df2.loc[l,'20~22 신규설치']=t
pivot_cctv_df2
```

```python
pivot_cctv_df2.loc['강원','20~22 신규설치']
```

```python
# 세종시에 설치 연도 데이터 X
# 지역별 총 카메라대수 추가

pivot_cctv_df3=pivot_cctv_df2.copy()
for l in local_list:
        idx=cctvDF[cctvDF['지역']==l].index
        j=pivot_cctv_df3.loc[l,'20~22 신규설치']
        t=cctvDF.loc[idx]['카메라대수'].sum()
        pivot_cctv_df3.loc[l,'19 총 카메라대수']=t-j
pivot_cctv_df3
```

```python
pivot_cctv_df3.loc['합계']=0
for i in pivot_cctv_df3.columns :
    pivot_cctv_df3.loc['합계',i]=pivot_cctv_df3[i].sum()
pivot_cctv_df3
```

```python
pivot_cctv_df3['20 카메라수']=0
pivot_cctv_df3['21 카메라수']=0
pivot_cctv_df3['22 카메라수']=0

for l in range(18):
    for y in [0,1,2]:
        j=pivot_cctv_df3.iloc[l,y]
        p=pivot_cctv_df3.iloc[l,4]
        if y==0:
            pivot_cctv_df3.iloc[l,y+5]=p+j
        elif y <= 2:
            pivot_cctv_df3.iloc[l,y+5]=pivot_cctv_df3.iloc[l,y+4]+j

        
pivot_cctv_df3
```

```python
pivot_cctv_df3['20 증가율']=0
pivot_cctv_df3['21 증가율']=0
pivot_cctv_df3['22 증가율']=0

for l in range(18):
    for i in range(4,7):
        a=pivot_cctv_df3.iloc[l,i]
        b=pivot_cctv_df3.iloc[l,i+1]
        if i <=6:
            pivot_cctv_df3.iloc[l,i+4]=(b-a)/a*100
pivot_cctv_df3
```
