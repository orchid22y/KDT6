```python
# 데이터 파일
TRAIN_PATH = '../data/lang/train/'
TEST_PATH='../data/lang/test/'

TRAIN_CSV='../data/lang_trian.csv'
TEST_CSV='../data/test/lang_test.csv'
```

```python
# 파일 읽기 => open()-> read(), readline(),readlines()->close()
#          => with open() as 구문 ==> with문만 빠져나가면 자동으로 close()

if not os.path.exists(TRAIN_CSV):
    print(f'{TRAIN_CSV}파일이 없습니다.')
else:
    with open(TRAIN_CSV, mode='r') as f : 
        alldata=f.read()
        print(f'[File Data] {alldata}')
```

```python
def make_dataset(dir_path,file_path):
    
    # 폴더 안에 파일 가져오기
    files=os.listdir(dir_path)  # os.listdir() : 파일 리스트를 문자열로 담아서 보여줌

    for file in files:
        print(file,dir_path+file, os.path.exists(dir_path+file))
        label=file[:2] #파일명에서 앞에 두글자만 가져오기 -> 라벨
        print(f'Label => {label}')
        with open(dir_path+file,mode='r') as f:
            #print(f'[File Data] {f.read()}')  # 데이터 읽기
            alldata=f.read()
        print(f'alldata => {alldata}')

        #대소문자 통일
        alldata=alldata.lower()
        print(f'alldata => {alldata}')
        print(f'alldata => {len(alldata)}')
        #알파벳이 아닌 문자 제거
        alldata=alldata.replace('\n','') #줄바꿈 제거 

        for ch in alldata:
            if ord('a') > ord(ch) or ord(ch) > ord('z'):
                alldata=alldata.replace(ch,'') # 알파벳이 아닌문자 제거
        print(f'alldata => {len(alldata)}')
        print(f'alldata => {alldata}')

        with open(file_path, mode='a') as f:
        # # - a ~ z 개수 카운팅 후 csv 파일 쓰기
        #     cc=dict(Counter(alldata)) # dict 형태로 저장
        #     f.write(label+',')  # '라벨,' 추가
        #     f.write(','.join([str(value) for value in cc.values()])+'\n') # 데이터 값 ','으로 연결하기

            #a~z 개수 카운팅 ==> 정렬, 전체 단어수로 나누기
            
            
            
        break
```
