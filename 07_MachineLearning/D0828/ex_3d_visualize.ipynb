```python
# 모듈로딩
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 데이터 생성
x=np.linspace(-10,10,11)
print(f'[x] {x.shape}\n{x}\n')

x=np.tile(x,(11,1))
print(f'[x] {x.shape}\n{x}\n')

y=np.transpose(x)
print(f'[y] {y.shape}\n{x}\n')

z= np.random.rand(11,11)
print(f'[z] {z.shape}\n{z}\n')
```

```python
# 시각화 진행
fig = plt.figure()

# 3차원 창 설정 => projection='3d'
ax=fig.add_subplot(111,projection='3d')

#평면 생성 : 2차원 어레이를 3차원  Surface로 표현
ax.plot_surface(x,y,z, color='yellow')
ax.set_zlim(-5,5)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```
